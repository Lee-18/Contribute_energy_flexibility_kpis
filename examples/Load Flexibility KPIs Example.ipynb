# An example for using the Python package

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import unittest
import sys
import os
```

## Load example data 

```python
parent_directory = os.path.dirname(os.path.dirname(os.path.abspath('Load Flexibility KPIs Example.ipynb')))
Data_directory = os.path.join(os.path.join(parent_directory, 'energy_flexibility_kpis'),'test')
DataSimple=pd.read_csv(os.path.join(Data_directory, 'data1_simple.csv'))

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(DataSimple['timestamp'],DataSimple['baseline_power'],label='base')
ax.plot(DataSimple['timestamp'],DataSimple['flexible_power'],label='shift')
ax.set_xlabel('timestamp')
ax.set_ylabel('power [W]')
ax.legend()
```

## Load example function

```python
parent_directory = os.path.dirname(os.path.dirname(os.path.abspath('Load Flexibility KPIs Example.ipynb')))
sys.path.append(parent_directory)

import kpi
```

## Calculate some KPIs using the example function and data

```python
result_peak_power_reduction = kpi.peak_power_reduction("1/1/2022 14:00",DataSimple['baseline_power'],DataSimple['flexible_power'],DataSimple['timestamp'])

result_flexibility_factor = kpi.flexibility_factor(DataSimple['baseline_power'],"2022-01-01 13:00:00","2022-01-01 15:00:00",DataSimple['timestamp'])

result_load_factor = kpi.load_factor(DataSimple['baseline_power'],DataSimple['timestamp'],"1/1/2022 14:00")

result_building_energy_flexibility_index = kpi.building_energy_flexibility_index(DataSimple['baseline_power'],DataSimple['flexible_power'])

print('peak_power_reduction=',result_peak_power_reduction, 'W')
print('flexibility_factor=',result_flexibility_factor)
print('load_factor=',result_load_factor)
print('building_energy_flexibility_index=',result_building_energy_flexibility_index)
```
